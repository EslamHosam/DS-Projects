# Guided Project Visualizing The Gender Gap In College Degrees

The Department of Education Statistics releases a data set annually containing the percentage of bachelor's degrees granted to women from 1970 to 2012. The data set is broken up into 17 categories of degrees, with each column as a separate category. **based on our dataset we are going to visualize the gender gap in our data for each category**
.
Randal Olson, a data scientist at University of Pennsylvania, has cleaned the data set and made it available on his personal website. You can [download the dataset here](http://www.randalolson.com/wp-content/uploads/percent-bachelors-degrees-women-usa.csv)

let's start by importing the dataset and important libraries

```python
# importing necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

#importing the dataset
women_degree = pd.read_csv("percent-bachelors-degrees-women-usa.csv")
women_degree.head()
```

## Ploting Details
now we need to give simple information for the gender gap in college categories via simple visualization for those categories
with some aesrhetics

### figure
we need to plot 17 graph one for each category so should have a figure with enough dimensions for all graphs it will be 
6 by 3 rectangle , will plot each list of categories in one column 

### subplots
to draw each list in one column we should take into consideration the number f each graph in the figure grid 
 [here is an image ](https://encrypted-tbn0.gstatic.com/images?q=tbn%3AANd9GcRc5Y3K1JuYPK-98O_wXGsl1l8XN5eLKEBEcw&usqp=CAU) to see an example for the index of each graph in (2,3) figure 
 
* so we have 6 X 3 rectangle (figure size) which is divided into 18 parts
* numering goes from left to right (6,3,1), (6,3,2)...(6,3,18)
* but notice that we have 3 columns so after the third one go to next row and so on 
* Our plan is to have subplots of each group in one column. So we need to access (6,3,1) (6,3,4) (6,3,7) and so on. So we have to construct our for loop accordingly. 

now lets go ahead and create our lists
* for aech male category percentage we need only to subtract the women percentage from 100

```python
# Creating STEM Category list

stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']

# Creating Liberal Category list

lib_arts_cats = [
    'Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 
    'Social Sciences and History']

# Creating Other Category list

other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

```python
# Creating Plotting Area
fig = plt.figure(figsize = (18,22))
fig.suptitle('Gender Gap in various Majors over the period of 1970 to 2010', fontsize=16)


# colour for each plot
cb_dark_blue = (0/255,107/255,164/255) # women
cb_orange = (255/255, 128/255, 14/255) # men

#  Start Create The Plot For Each List
# STEM GROUP
for i in range(0,len(stem_cats)):
    ax = fig.add_subplot(6,3, (i*3 +1))
    ax.plot(women_degree["Year"], women_degree[stem_cats[i]], linewidth= 2.5, c= cb_dark_blue)
    ax.plot(women_degree["Year"], 100- women_degree[stem_cats[i]], linewidth= 2.5, c= cb_orange)
    
    # set xlim and ylim
    ax.set_xlim(1968,2011)
    ax.set_ylim(0,100)
    
    #set title for each graph
    ax.set_title(stem_cats[i])
    
    # hiding ticks and hide the x- axis labels
    ax.tick_params(left= False, right= False, bottom= False, top= False, labelbottom = False )
    
    # hiding spines
    for key, spine in ax.spines.items():
        spine.set_visible(False)
    
    # seting y_ticks to show which value to be displayed on y axis
    ax.set_yticks([0,100])
    
    # adding labels for easy understanding
    if i == 0:
        ax.text(2009, 12, "Men")
        ax.text(2011, 86, "Women")
    elif i == 5:
        ax.text(2011, 12, "Women")
        ax.text(2011, 70, "Men")
        #enable the x-axis label for the last graph only
        ax.tick_params(labelbottom = True)
        
    # Adding horizental line between the 2 plots
    ax.axhline(50 , c = (171/255, 171/255, 171/255), alpha = 0.3)
    
    
# we will plot other categories with bit differences between them
# Plot lib_arts_cats 
for i in range(0, len(lib_arts_cats)):
    ax= fig.add_subplot(6,3, (3*i +2))
    ax.plot(women_degree["Year"], women_degree[lib_arts_cats[i]], linewidth= 2.5, c=cb_dark_blue)
    ax.plot(women_degree["Year"], 100 - women_degree[lib_arts_cats[i]], linewidth= 2.5, c= cb_orange)
    
    # hide ticks and x-axis labels
    ax.tick_params(left= False, top= False, right= False,bottom= False, labelbottom= False)
    
    # hide the spines
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["right"].set_visible(False)
    
    # set y labels to (0,100)
    ax.set_yticks([0,100])
    
    # add labels to the plots
    if i == 0:
        ax.text(2011, 25, "Men")
        ax.text(2011, 85, "Women")
    elif i == 4:
        ax.text(1980, 20, "Women")
        ax.text(1980, 70, "Men")
        
        # add x-axis labels to last plot
        ax.tick_params(labelbottom = True)
        
    # add horizental line between plots
    ax.axhline(50, c = (171/255, 171/255, 171/255) , alpha = 0.3)
    
    # adding titles
    ax.set_title(lib_arts_cats[i])

# Plot other cats 
for i in range(0, len(other_cats)):
    ax= fig.add_subplot(6,3, (3*i +3))
    ax.plot(women_degree["Year"], women_degree[other_cats[i]], linewidth= 2.5, c=cb_dark_blue)
    ax.plot(women_degree["Year"], 100 - women_degree[other_cats[i]], linewidth= 2.5, c= cb_orange)
    
    # hide ticks and x-axis labels
    ax.tick_params(left= False, top= False, right= False,bottom= False, labelbottom= False)
    
    # hide the spines
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)
    ax.spines["bottom"].set_visible(False)
    ax.spines["right"].set_visible(False)
    
    # set y labels to (0,100)
    ax.set_yticks([0,100])
    
    # add labels to the plots
    if i == 0:
        ax.text(2011, 25, "Men")
        ax.text(2011, 85, "Women")
    elif i == 5:
        ax.text(1985, 20, "Women")
        ax.text(1985, 70, "Men")
        
        # add x-axis labels to last plot
        ax.tick_params(labelbottom = True)
        
    # add horizental line between plots
    ax.axhline(50, c = (171/255, 171/255, 171/255) , alpha = 0.3)
    
    # adding titles
    ax.set_title(other_cats[i])
    
# FINAL STEP IS TO SAVE YOUR GRAPH AS A FILE
#SAVE AS AN IMAGE using pyplot.savefig("give image name") before pyplot.show

plt.savefig("Visualizing The Gender Gap In College Degrees")
plt.show()
```

In 12 out of 18 majors we have analyses, women are either ahead of men or in neck to neck competition.
